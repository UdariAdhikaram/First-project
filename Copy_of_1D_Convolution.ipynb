```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal = np.array([1, 1, 1, 1, 1, 1, 1, 0.5, 0.5, 0.5])
```

```python
plt.stem(signal)
plt.show()
```

```python
filter = np.array([-1, 1])
```

```python
result = np.convolve(signal, filter, 'valid')
```

```python
plt.stem(result)
plt.show()
```
